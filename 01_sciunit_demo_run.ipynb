## User 1: Creating a Sciunit (Hello World)

In this section, **User 1** will act as the experiment author.

We will:
1. Create a new Sciunit container
2. Write a simple Hello World shell script
3. Capture the execution using `sciunit exec`

Open **Terminal 1** and follow the steps below.


# Ensure we are User 1
whoami

# Create a new Sciunit container
sciunit create Project1


Expected output: 
Opened empty Sciunit container at ~/sciunit/Project1


### Create a Hello World program

Now we create a simple shell script that prints "hello, world".


# Create hello.sh
write the following commands one by one in terminal and enter.

cat > hello.sh << 'EOF'

#!/bin/sh

echo "hello, world"

EOF

Now hello world has been added to script hello.sh, if you want to verify type the optional command 
nano hello.sh 
you should see (#!/bin/sh, hello, world) 
once verfied use ctrl + x to exit the script.

# Make it executable
chmod +x hello.sh

# Run it normally
./hello.sh


Expected output: 
hello, world


### Capture the execution with Sciunit

Now we rerun the program using `sciunit exec`.
This captures the command **and all runtime dependencies**.


# Capture the execution
sciunit exec ./hello.sh


Expected output: hello, world


Sciunit has now captured this execution as **execution e1**.

We can inspect the captured execution.


# Show details of the captured execution
sciunit show e1


At this point, **User 1 has created a Sciunit container** with one execution (`e1`).

Next, we will share this Sciunit with User 2 so it can be reproduced.


### Share the Sciunit container

User 1 now packages the Sciunit container so it can be used by another user.


# Package the Sciunit container

cd ~

tar -czf Project1_sciunit.tar.gz sciunit


## User 2: Reproducing the Sciunit

User 2 represents an independent reproducer.

Open **Terminal 2**, switch to user2, and follow the steps below.


# Become User 2

whoami

Expected output: user2


### Import the Sciunit container from User 1

User 2 does not create the experiment.
Instead, User 2 receives the Sciunit container created by User 1.


# Copy the Sciunit archive from User 1's home
cp /home/nkmh5_missouri_edu/Project1_sciunit.tar.gz ~

# Extract it
tar -xzf Project1_sciunit.tar.gz

this creates ~/sciunit/Project1

(optional) if you want to confirm and see the extracted output execute the below command

ls 

you should see "sciunit" directory in it.




### Open the Sciunit container

User 2 now opens the Sciunit container created by User 1.


# Open the Sciunit container
sciunit open Project1

# List available executions
sciunit list


Expected output: 

e1  ./hello.sh


### Repeat the captured execution

User 2 now repeats execution `e1` without re-running the original script.


# Repeat execution e1
sciunit repeat e1

Expected output: hello, world


## What did we demonstrate?

- User 1 created and captured an execution using Sciunit
- Sciunit recorded all runtime dependencies
- User 2 received only the Sciunit container
- User 2 reproduced the execution using `sciunit repeat`

This demonstrates **reproducibility using Sciunit**.
