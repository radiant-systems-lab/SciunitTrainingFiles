# Ubuntu VM: Creating a POMD sciunit container

In this section, **Ubuntu user** will act as the experiment author.

We will:
1. Create a new Sciunit container
2. Execute the POMD file
3. copy the experiment to use it to run in CentOS machine.

Open **Ubuntu Terminal** and follow the steps below.



Before creating a container lets ssh using the command shown in "00_environment_check jupyter notebook"

it should be something like 
"ssh -i ~/ssh_keys/P_Key_NASA_DEMO.pem cc@129.114.xx.xxx"
if prompted a yes/no question, type yes and enter.

you should now be in ubuntu machine

### Move to required directory
lets move to directory where the files are present to execute 

cd SciunitTrainingFiles

you will have the POMD file and Data files required to run the experiment

## Create a new Sciunit container
sciunit create Project


Expected output: 
Opened empty Sciunit container at ~/sciunit/Project


## Execute the POMD file

sciunit exec python3 POMDFile_lite.py

The sciunit will start to execute and show the progress. 
once done it shows script executed successfully.



## Copy the sciunit container

sciunit copy

This should provide you a link which can be used to repeat in cent OS machine.
copy it.

ex: "https://d3okuktvxs1y4w.cloudfront.net/2026-01-29-13:51:15//home/cc/sciunit/project.zip"


# Cent OS VM: Repeating POMD sciunit container

In this section, **CentOS user** will act as the experiment reproducer.

We will:
1. Open the copy
2. Repeat the POMD file

Open **Cent OS Terminal** and follow the steps below.

lets ssh using the command shown in "00_environment_check jupyter notebook"

it should be something like 
"ssh -i ~/ssh_keys/P_Key_NASA_DEMO.pem cc@129.114.xx.xxx"
if prompted a yes/no question, type yes and enter.

you should now be in cent OS machine

## Open the container
Use the following command and link which we copied in ubuntu machine, we will open the project

sciunit open link

place link at link section, once done executing command, you should see "Switched to sciunit 'project"

## (optional)

if you want to check whether we have the required  experiment e1 use the following commands

sciunit list 

this should show you list of experiment present in the project we just opened. 
ex:  e1 Jan 29 06:48 python3 POMDFile_lite.py


## Repeat the POMD experiment 

sciunit repeat e1

sciunit should take less than minute to execute and show you the repeated steps

you have now successfully reproduced the same experiment on cross platform.
